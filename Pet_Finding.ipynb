```python
#import the neccesary libraries.
#the petpy library helps you interact with petfinder API. You can read more on how to download/use it here: https://petpy.readthedocs.io/en/latest/
import petpy
import numpy
import json
import time
import pandas as pd
```

```python
#rename the key/token to be able to use Petfinder's API
key = PETFINDER_KEY
secret = PETFINDER_SECRET_KEY
```

```python
#send a request
pf = petpy.Petfinder(key=key, secret=secret)
```

```python
#Check if you get a response. If not do not precede foward and be sure to check if the keys are correct and correctly named.
pf
```

```python
animal_base = pf.animals()
animal_base
```

```python
animal_types = pf.animal_types()
```

```python
#check possible animal options and their coats
for animal in animal_types['types']:
    print(animal['name'], '\n', animal['coats'])
```

```python
#Now lets go through a run through of finding an adorable animal to adopt!
```

```python
#I'm requesting to find animals that have these characteristics to be true: Adoptable and near Washington DC. 
#These can be altered and different traits can be added
close_adopts_df = pf.animals(status='adoptable', location='Washington, DC', distance=10, pages=None, return_df=True)
close_adopts_df
```

```python
#There are alot of columns, but I don't think I'll need all that information in the future.
#This will show us each column title and to figure out which ones we would like to keep later on.
for animal in close_adopts_df:
    print(animal)
```

```python
#using the list we just created, I would like to know all the unique pets names
names_pet = close_adopts_df["name"].unique()
#then I would like to sort them alphabetically
names_pet.sort()
names_pet
```

```python
#For fun lets look at the most common names from this list
common_name = close_adopts_df['name'].value_counts()[:10].sort_values(ascending=False)
common_name
```

```python
#I would also like to know what breeds are availble to be adopted
breeds_pet = close_adopts_df['breeds.primary'].unique()
breeds_pet.sort()
breeds_pet
```

```python
#Besides cats or dogs, I would like to see if there are any other types of adoptable animals
common_animals = close_adopts_df['species'].value_counts()[:10].sort_values(ascending=False)
common_animals
```

```python
#Some places near DC are harder to get to than others, so I just want to take a look at the subset of areas
towns = close_adopts_df['contact.address.city'].unique()
towns
```

```python
#Now I've created a list of what the columns I'm interested in
DC_adoptable = close_adopts_df[['name','species', 'breeds.primary','age', 'gender', 'size', 'coat', 'description','contact.phone', 'contact.address.city', 'contact.address.state', 'contact.address.postcode', 'url', 'primary_photo_cropped.large']].copy()
DC_adoptable
```

```python
#Using all the information we have lets put down some criteria of what animals that would be a perfect fit for me!
#I'm looking for an old, large, and female cat. Now I'll take this criteria and see if this narrows down the list.
my_pet = DC_adoptable.loc[(DC_adoptable['species'] == 'Cat') & (DC_adoptable['age'] == 'Senior') & (DC_adoptable['gender'] == 'Female') & (DC_adoptable['size'].str.contains('Large'))]
my_pet
```

```python
#Lets create a csv from the small list and an excel document from the larger list, so it easy information to share with other people!
my_pet.to_csv("final_pets.csv", encoding="utf-8", index=False)
DC_adoptable.to_excel('dc_area_pets.xlsx', sheet_name='Pets_Ten_Miles', index=False)
```
